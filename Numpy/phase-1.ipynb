### LIST VS NUMPY ARRAY

```python

py_list = [1,2,3]
print("Python list multiplication ",py_list*2)
py_array = np.array([1,2,3])
print("Python list multiplication ",py_array*2)
```

#### _Numpy Arrays just work like matrices_

```python
import time
start_time = time.time()
py_list = [i*2 for i in range(100000000)]
print("\nList Operation time: ",time.time()-start_time)

start_time = time.time()
np_array = np.arange(100000000) * 2
print("\nArray Operation time: ",time.time()-start_time)
```
